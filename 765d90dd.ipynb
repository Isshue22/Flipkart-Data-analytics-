# Flipkart Data Analysis

This notebook shows a simple end-to-end Flipkart data analysis using pandas, numpy, Matplotlib, and Seaborn.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For pretty charts
sns.set_style('whitegrid')
```

## 1. Create Sample Flipkart DataFrame

```python
# Create a simple random Flipkart-style dataset.
np.random.seed(7)
categories = ['Electronics', 'Clothing', 'Books', 'Home', 'Beauty']
brands = ['Samsung', 'Apple', 'Nike', 'Sony', 'LG', 'MI']

N = 200
df = pd.DataFrame({
    'Category': np.random.choice(categories, N),
    'Brand': np.random.choice(brands, N),
    'Rating': np.round(np.random.uniform(3, 5, N), 1),
    'Price': np.random.randint(500, 20000, N),
    'Discount(%)': np.random.randint(5, 70, N),
    'Reviews': np.random.randint(0, 5000, N)
})
# Show a sample
df.head()
```

## 2. Data Info and Missing Values

```python
# Check data info and nulls
print(df.info())
print(df.isnull().sum())
```

## 3. Category Distribution (Bar Plot)

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Category', palette='pastel')
plt.title('Product Count by Category')
plt.show()
```

## 4. Brand Distribution (Pie Chart)

```python
plt.figure(figsize=(5,5))
df['Brand'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title('Brand Share')
plt.ylabel('')
plt.show()
```

## 5. Price Distribution (Histogram)

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Price'], bins=20, kde=True, color='teal')
plt.title('Product Price Distribution')
plt.xlabel('Price (INR)')
plt.show()
```

## 6. Discount vs Price (Scatter Plot)

```python
plt.figure(figsize=(7,4))
sns.scatterplot(data=df, x='Discount(%)', y='Price', hue='Category', alpha=0.8)
plt.title('Discount vs Price')
plt.show()
```

## 7. Rating by Category (Box Plot)

```python
plt.figure(figsize=(7,4))
sns.boxplot(data=df, x='Category', y='Rating', palette='Blues')
plt.title('Rating by Category')
plt.show()
```

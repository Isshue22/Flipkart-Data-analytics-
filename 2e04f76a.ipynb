```python
# 1. Import Libraries and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime, timedelta

# Configure display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('default')
sns.set_palette("husl")

print("📚 Libraries imported successfully!")
```

## 2. Dataset Generation and Loading

We'll create a realistic Flipkart dataset with various product categories, brands, and pricing information.

```python
# Generate realistic Flipkart dataset
np.random.seed(42)

# Define categories and brands
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Books', 
             'Beauty & Personal Care', 'Sports & Fitness', 'Automotive', 
             'Mobiles & Accessories']

brands = ['Samsung', 'Apple', 'Nike', 'Adidas', 'Sony', 'LG', 'HP', 
         'Dell', 'Lenovo', 'MI', 'Realme', 'OnePlus', 'Boat', 'JBL', 
         'Puma', 'Reebok', 'Philips', 'Bajaj', 'Prestige']

n_products = 2000

# Create comprehensive dataset
data = {
    'product_id': ['FK_' + str(i+1).zfill(6) for i in range(n_products)],
    'product_name': [f'Product {i+1}' for i in range(n_products)],
    'category': np.random.choice(categories, n_products),
    'brand': np.random.choice(brands, n_products),
    'retail_price': np.random.uniform(500, 50000, n_products).round(2),
    'discount_percentage': np.random.uniform(5, 70, n_products).round(2),
    'rating': np.random.choice([0, 3.5, 4.0, 4.2, 4.5, 4.8, 5.0], 
                              n_products, p=[0.2, 0.1, 0.15, 0.2, 0.2, 0.1, 0.05]),
    'reviews_count': np.random.randint(0, 5000, n_products),
    'seller_name': np.random.choice(['Cloudtail India', 'Appario Retail', 
                                   'RetailNet', 'SuperComNet'], n_products),
    'is_flipkart_assured': np.random.choice([True, False], n_products, p=[0.3, 0.7]),
    'availability': np.random.choice(['In Stock', 'Out of Stock', 'Limited Stock'], 
                                   n_products, p=[0.7, 0.1, 0.2])
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate selling price based on discount
df['selling_price'] = (df['retail_price'] * (1 - df['discount_percentage']/100)).round(2)

# Add timestamp column
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 12, 31)
df['crawl_timestamp'] = [start_date + timedelta(days=np.random.randint(0, (end_date - start_date).days)) 
                        for _ in range(n_products)]

print(f"✅ Dataset created successfully with {df.shape[0]} rows and {df.shape[1]} columns")
print(f"Memory usage: {df.memory_usage().sum() / 1024:.2f} KB")
```

## 3. Initial Data Exploration

```python
# Basic dataset information
print("📊 DATASET OVERVIEW")
print("=" * 50)

# Display first few rows
print("\n🔍 First 5 rows:")
display(df.head())

# Dataset info
print("\n📋 Dataset Info:")
print(df.info())

# Statistical summary
print("\n📈 Statistical Summary:")
display(df.describe())
```

## 4. Data Preprocessing and Feature Engineering

```python
print("🔧 DATA PREPROCESSING")
print("=" * 40)

# Handle missing values
df['brand'].fillna('Unknown', inplace=True)
df['rating'] = df['rating'].replace(0, np.nan)

# Feature Engineering
# Price categories
df['price_category'] = pd.cut(df['selling_price'], 
                             bins=[0, 1000, 5000, 15000, 50000], 
                             labels=['Budget', 'Mid-Range', 'Premium', 'Luxury'])

# Rating categories  
df['rating_category'] = pd.cut(df['rating'], 
                              bins=[0, 3, 4, 4.5, 5], 
                              labels=['Poor', 'Average', 'Good', 'Excellent'])

# Calculate savings
df['savings'] = df['retail_price'] - df['selling_price']

# Extract temporal features
df['year'] = df['crawl_timestamp'].dt.year
df['month'] = df['crawl_timestamp'].dt.month
df['quarter'] = df['crawl_timestamp'].dt.quarter

print("✅ Feature engineering completed!")
print(f"New shape: {df.shape}")

# Save processed dataset
df.to_csv('flipkart_products_dataset.csv', index=False)
print("💾 Dataset saved as 'flipkart_products_dataset.csv'")
```

## 5. Exploratory Data Analysis (EDA)

```python
print("🔍 EXPLORATORY DATA ANALYSIS")
print("=" * 45)

# Missing values analysis
print("❓ Missing Values:")
print(df.isnull().sum())

# Category distribution
print("\n📊 Category Distribution:")
print(df['category'].value_counts())

# Price analysis
print(f"\n💰 Price Analysis:")
print(f"Average selling price: ₹{df['selling_price'].mean():,.2f}")
print(f"Median selling price: ₹{df['selling_price'].median():,.2f}")
print(f"Price range: ₹{df['selling_price'].min():,.2f} - ₹{df['selling_price'].max():,.2f}")

# Discount analysis
print(f"\n🏷️ Discount Analysis:")
print(f"Average discount: {df['discount_percentage'].mean():.1f}%")
print(f"Maximum discount: {df['discount_percentage'].max():.1f}%")
print(f"Total savings: ₹{df['savings'].sum():,.2f}")
```

## 6. Advanced Analytics

```python
# Category-wise analysis
print("📋 CATEGORY-WISE ANALYSIS")
print("=" * 35)

category_analysis = df.groupby('category').agg({
    'selling_price': ['mean', 'median', 'min', 'max'],
    'discount_percentage': 'mean',
    'rating': 'mean',
    'reviews_count': 'mean',
    'savings': 'mean'
}).round(2)

display(category_analysis)

# Brand performance analysis
print("\n🏷️ BRAND PERFORMANCE ANALYSIS")
print("=" * 40)

brand_analysis = df.groupby('brand').agg({
    'selling_price': 'mean',
    'discount_percentage': 'mean', 
    'rating': 'mean',
    'reviews_count': 'mean',
    'product_id': 'count'
}).round(2).rename(columns={'product_id': 'product_count'})

top_brands = brand_analysis.sort_values('rating', ascending=False).head(10)
print("Top 10 Brands by Rating:")
display(top_brands)

# Seller analysis
seller_performance = df.groupby('seller_name').agg({
    'selling_price': 'mean',
    'discount_percentage': 'mean',
    'rating': 'mean',
    'product_id': 'count',
    'is_flipkart_assured': lambda x: (x == True).sum() / len(x) * 100
}).round(2)

print("\n🏪 Seller Performance:")
display(seller_performance)
```

## 7. Data Visualization Dashboard

```python
# Set up the plotting environment
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 10

# Create comprehensive visualizations
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Flipkart E-commerce Data Analysis Dashboard', fontsize=16, fontweight='bold')

# 1. Category Distribution
df['category'].value_counts().plot(kind='bar', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Product Distribution by Category')
axes[0,0].set_xlabel('Category')
axes[0,0].set_ylabel('Count')
axes[0,0].tick_params(axis='x', rotation=45)

# 2. Price Distribution
df['selling_price'].hist(bins=30, ax=axes[0,1], color='lightgreen', alpha=0.7)
axes[0,1].set_title('Selling Price Distribution')
axes[0,1].set_xlabel('Selling Price (₹)')
axes[0,1].set_ylabel('Frequency')

# 3. Rating Distribution
df['rating'].value_counts().sort_index().plot(kind='bar', ax=axes[0,2], color='orange')
axes[0,2].set_title('Customer Rating Distribution')
axes[0,2].set_xlabel('Rating')
axes[0,2].set_ylabel('Count') 

# 4. Discount vs Savings Scatter Plot
scatter = axes[1,0].scatter(df['discount_percentage'], df['savings'], 
                           c=df['selling_price'], cmap='viridis', alpha=0.6)
axes[1,0].set_title('Discount % vs Savings Amount')
axes[1,0].set_xlabel('Discount Percentage (%)')
axes[1,0].set_ylabel('Savings (₹)')
plt.colorbar(scatter, ax=axes[1,0], label='Selling Price (₹)')

# 5. Category-wise Price Box Plot
df.boxplot(column='selling_price', by='category', ax=axes[1,1])
axes[1,1].set_title('Price Distribution by Category')
axes[1,1].set_xlabel('Category')
axes[1,1].set_ylabel('Selling Price (₹)')

# 6. Top Brands by Average Rating
top_10_brands = brand_analysis.sort_values('rating', ascending=False).head(10)
top_10_brands['rating'].plot(kind='bar', ax=axes[1,2], color='purple')
axes[1,2].set_title('Top 10 Brands by Average Rating')
axes[1,2].set_xlabel('Brand')
axes[1,2].set_ylabel('Average Rating')

plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
numerical_cols = ['retail_price', 'selling_price', 'discount_percentage', 
                 'rating', 'reviews_count', 'savings']
correlation_matrix = df[numerical_cols].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix of Key Metrics', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

# Flipkart Data Analysis

This notebook answers 6 key data questions about Flipkart product listings.

## Questions Answered

- 1. Which product categories are the most and least common?
- 2. What is the price distribution for products on Flipkart?
- 3. Which brands appear most frequently?
- 4. Is there a relation between discount percentage and price?
- 5. How are product ratings distributed by category?
- 6. Which seller has the most products and best average rating?

We use pandas for data analysis and seaborn/matplotlib for visualization. The data is read from the CSV generated in the first step.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
# Read the data
df = pd.read_csv('flipkart_products_dataset.csv')
df.head()
```

## 1. Which product categories are the most and least common?

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='category', order=df['category'].value_counts().index, palette='pastel')
plt.title('Product Count by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.show()

print('Most common:', df["category"].value_counts().idxmax())
print('Least common:', df["category"].value_counts().idxmin())
```

## 2. What is the price distribution for products on Flipkart?

```python
plt.figure(figsize=(8,4))
sns.histplot(df['selling_price'], bins=30, kde=True, color='skyblue')
plt.title('Selling Price Distribution')
plt.xlabel('Selling Price (INR)')
plt.show()
```

## 3. Which brands appear most frequently?

```python
top_brands = df['brand'].value_counts().head(10)
plt.figure(figsize=(8,4))
sns.barplot(x=top_brands.values, y=top_brands.index, palette='pastel')
plt.title('Top 10 Most Frequent Brands')
plt.xlabel('Count')
plt.ylabel('Brand')
plt.show()
```

## 4. Is there a relation between discount percentage and price?

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='discount_percentage', y='selling_price', alpha=0.5)
plt.title('Discount % vs Selling Price')
plt.xlabel('Discount (%)')
plt.ylabel('Selling Price (INR)')
plt.show()

# Optional: Calculate correlation
print('Correlation:', df[['discount_percentage', 'selling_price']].corr().iloc[0,1].round(2))
```

## 5. How are product ratings distributed by category?

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='category', y='rating', palette='light:#5A9')
plt.title('Rating Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```

## 6. Which seller has the most products and best average rating?

```python
seller_counts = df['seller_name'].value_counts()
top_seller = seller_counts.idxmax()
best_seller = df.groupby('seller_name')['rating'].mean().idxmax()

print('Seller with most products:', top_seller)
print('Seller with best average rating:', best_seller)
```
